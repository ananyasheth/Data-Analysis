```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from basic_units import cm, inch

df = pd.read_excel("IPL.xls")
#print(df)

l1 = []
l2 = []
l3 = []
l4 = []
l5 = []
l6 = []

#To find the team having the highest average winning margin (seperately for runs and wickets)
df1= df['winner'].value_counts()
df2 = pd.DataFrame(df1)
df2 = df2.reset_index()
df2 = df2.rename(columns={'index':'Winner'})
df2 = df2.rename(columns={'winner' : 'Matches'})

for i in range(0,df2.shape[0]):
    l1.append(df2.iloc[i,0])
    l6.append(df2.iloc[i,1])
    
for j in range(0, len(l1)):
    df3 = df[(df['winner']== l1[j]) & (df['Win_Type'] == 'run')]
    df4 = pd.DataFrame(df3['Winning_Margin'])
    df4 = df4.reset_index()
    
    for k in range(0,df4.shape[0]):
        l2.append(df4.iloc[k,1])
    
        
    l3.append(l2)
    l2 = []


for h in range(0,len(l3)):
    sum = 0
    Average = 0
    for f in range(0, len(l3[h])):
        sum = l3[h][f] + sum
    
    if len(l3[h])!= 0:
        Average = sum/len(l3[h])
    l5.append(Average)

N = len(l5)

fig, ax = plt.subplots(figsize = (10,6))

ind = np.arange(N)   
width = 0.35   
    
p1 = ax.bar(ind, l6, width, bottom=0*cm)
p2 = ax.bar(ind + width, l5, width, bottom=0*cm)

ax.set_title('IPL Data Analysis')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(l1)

ax.legend((p1[0], p2[0]), ('Matches Won', 'Average winning margin'))
ax.yaxis.set_units(inch)
ax.autoscale_view()


plt.show()
```
